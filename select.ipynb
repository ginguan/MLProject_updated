#  This ipynb is used for Data preprocessing

```python
import pandas as pd
from pandas_summary import DataFrameSummary
import matplotlib.pyplot as plt
bid = pd.read_csv("select.csv") #this csv file include the balanced/joint dataset from bid.csv and train.csv
bid.head()
```

# determine if missing value observed

```python

bid = bid.dropna()
pd.isnull(bid).any()
```

# normalization

```python
#normalize "time" column
bid["time"]=((bid["time"]-bid["time"].min())/(bid["time"].max()-bid["time"].min()))*100
#combine.sample(5)
```

```python
# change all device data to numerical value by subtracting "phone" in string
bid['device']=bid['device'].str.strip('phone')
bid['device']=bid['device'].astype(int)
bid.sample(5)
```

#  

```python
#Rank and normalize all values from "merchandise" to range [0.1,1]
bid['merchandise']=bid['merchandise'].str.replace('books and music','0.1')
bid['merchandise']=bid['merchandise'].str.replace('office equipment','0.2')
bid['merchandise']=bid['merchandise'].str.replace('computers','0.3')
bid['merchandise']=bid['merchandise'].str.replace('clothing','0.4')
bid['merchandise']=bid['merchandise'].str.replace('home goods','0.5')
bid['merchandise']=bid['merchandise'].str.replace('furniture','0.6')
bid['merchandise']=bid['merchandise'].str.replace('sporting goods','0.7')
bid['merchandise']=bid['merchandise'].str.replace('auto parts','0.8')
bid['merchandise']=bid['merchandise'].str.replace('mobile','0.9')
bid['merchandise']=bid['merchandise'].str.replace('jewelry','1.0')
merchandise=bid[ "merchandise" ].unique( )
print(merchandise) # showing all categories after processed in "merchandise" colum
```

```python

#check how many different/uniqe values in each features
bidder_id=bid[ "bidder_id" ].unique( )
print('bidder_id',len(bidder_id))
bid_id=bid[ "bid_id" ].unique( )
print('bid_id',len(bid_id))
auction=bid[ "auction" ].unique( )
print('auction',len(auction))
merchandise=bid[ "merchandise" ].unique( )
print('merchandise',len(merchandise))
device=bid[ "device" ].unique( )
print('device',len(device))
country=bid[ "country" ].unique( )
print('country',len(country))
url=bid[ "url" ].unique( )
print('url',len(url))
payment_account=bid[ "payment_account" ].unique( )
print('payment_account',len(payment_account))
address=bid[ "address" ].unique( )
print('address',len(address))
```

```python
#beacause now bid table has a defaut header column, in order to get rid of that index column
#transfer to a new csv file "select2.csv"
#also because the payment_account and address are detected same when along with bidder_id, therefore, 
#these 2 columns are dropped when transfering
header =['bid_id','bidder_id','auction','merchandise','device','time','country','ip','url','outcome']
bid['outcome']=bid['outcome'].astype(int)
bid.to_csv('select2.csv',columns = header,index=False)
```

```python
select = pd.read_csv("select2.csv")
select.head()
```

# Normalze data using LabelEncoder from scikitlearn. preprocessing library 

```python
from sklearn.preprocessing import LabelEncoder
country=select[ "country" ].unique( )
#print(source_labels.str.sort())
print(len(country))
label = LabelEncoder()
select[ "country" ] = label.fit_transform(select[ "country" ])
select.head()
```

```python
bidder_id=select[ "bidder_id" ].unique( )
print(len(bidder_id))
label = LabelEncoder()
select[ "bidder_id" ] = label.fit_transform(select[ "bidder_id" ])
select.head()
```

```python
#bidder_id=select[ "auction" ].unique( )
#print(len(bidder_id))
label = LabelEncoder()
select[ "auction" ] = label.fit_transform(select[ "auction" ])
select.head()
```

```python
#url=select[ "url" ].unique( )
#print(len(url))
label = LabelEncoder()
select[ "url" ] = label.fit_transform(select[ "url" ])
select.head()
```

```python
#ip=select[ "ip" ].unique( )
label = LabelEncoder()
select[ "ip" ] = label.fit_transform(select[ "ip" ])
select.head()
```

# Set Scaling in columns to reduce problems occur with range 


```python
from sklearn import preprocessing
from sklearn.preprocessing import MinMaxScaler
temp = select[['auction']].values
scaler = preprocessing.MinMaxScaler(feature_range=(0,100))
scaled_values = scaler.fit_transform(temp) 
nornalized = pd.DataFrame(scaled_values)
select['auction'] = nornalized

temp = select[['device']].values
scaler = preprocessing.MinMaxScaler(feature_range=(0,100))
scaled_values = scaler.fit_transform(temp) 
nornalized = pd.DataFrame(scaled_values)
select['device'] = nornalized



temp = select[['ip']].values
scaler = preprocessing.MinMaxScaler(feature_range=(0,100))
scaled_values = scaler.fit_transform(temp) 
nornalized = pd.DataFrame(scaled_values)
select['ip'] = nornalized

temp = select[['url']].values
scaler = preprocessing.MinMaxScaler(feature_range=(0,100))
scaled_values = scaler.fit_transform(temp) 
nornalized = pd.DataFrame(scaled_values)
select['url'] = nornalized



select.head()
```

# Write a csv file as selectResult.csv for result of processed data

```python

header =['bid_id','bidder_id','auction','merchandise','device','time','country','ip','url','outcome']

select.to_csv('selectResult.csv',columns = header,index=False)
select.head()
```
